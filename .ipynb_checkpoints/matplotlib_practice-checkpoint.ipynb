```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of stings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", " Sept", "Oct", "Nov", "Dec"]

# Set y-axis to a list of floats as the total fare in the US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 27.04, 43.82, 10.56, 11.85, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
```

```python
# Create the plt with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plt with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plt with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot 
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')

#Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45 
plt.ylim(0,45)

# Create a title 
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```
